```python
def solution(progresses, speeds):
    import numpy as np
    
    progresses = np.array(progresses)
    speeds = np.array(speeds)
    a = len(progresses)
    answer= []
    while len(progresses)>(progresses>=100).sum():
        progresses = progresses + speeds[a-len(progresses):]
#         print(progresses>=100)

        max_sum=[0]
        for i,j in enumerate(list(progresses>=100)):
#             print(i+1,((progresses>=100)[:i+1]))
            if i+1 == ((progresses>=100)[:i+1]).sum():
                max_sum.append(((progresses>=100)[:i+1]).sum())

        b = max(max_sum)        
        answer.append(b)
        progresses = progresses[b:]
    answer = [x  for x in list(answer) if x >= 1]
    return answer
```

위를 실행하면 값이 나오지만, 
`TypeError: Object of type 'int64' is not JSON serializable`
가 나오게 된다. 

해결 방법은 https://frhyme.github.io/python-libs/int64_is_not_json_serializable/
* int64 형을 int 로 다시 변환 해주면 된다. 

* json 값은 int64를 인식 못하기 때문에

아래와 같이 int를 적용해 주면 끝~

```python
def solution(progresses, speeds):
    import numpy as np
    
    progresses = np.array(progresses)
    speeds = np.array(speeds)
    a = len(progresses)
    answer= []
    while len(progresses)>(progresses>=100).sum():
        progresses = progresses + speeds[a-len(progresses):]
#         print(progresses>=100)

        max_sum=[0]
        for i,j in enumerate(list(progresses>=100)):
#             print(i+1,((progresses>=100)[:i+1]))
            if i+1 == ((progresses>=100)[:i+1]).sum():
                max_sum.append(((progresses>=100)[:i+1]).sum())

        b = max(max_sum)        
        answer.append(b)
        progresses = progresses[b:]
    answer = [int(x)  for x in list(answer) if x >= 1]
    return answer
```

```python
def solution(progresses, speeds):
    print(progresses)
    print(speeds)
    answer = []
    time = 0
    count = 0
    while len(progresses)> 0:
        if (progresses[0] + time*speeds[0]) >= 100:
            progresses.pop(0)
            speeds.pop(0)
            count += 1
        else:
            if count > 0:
                answer.append(count)
                count = 0
            time += 1
    answer.append(count)
    return answer
```
