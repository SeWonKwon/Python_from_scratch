```python
def solution(lottos, win_nums):
    nums_0 = 0
    for i in lottos:
        if i == 0:
            nums_0+=1
    lottos = set(lottos)
    win_nums = set(win_nums)
    matched_num = len(lottos & win_nums)
    ranklist = [6, 6, 5, 4, 3, 2, 1]
    return ranklist[matched_num+nums_0],ranklist[matched_num]
```

```python
lottos=[45, 4, 35, 20, 3, 9]
win_nums=[20, 9, 3, 45, 4, 35]
print(solution(lottos, win_nums))
```
