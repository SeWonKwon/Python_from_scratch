짝수 : 2로 나누고   
홀수 : 3을 곱하고 + 1      
500 번 넘어가면 리턴 -1     

n	result
6	8
16	4
626331	-1

```python
n = 6
```

```python
def solution(n):
    cnt = 0
    while True:
        if n == 1 : 
            return cnt
        elif cnt == 500: 
            return -1
        elif n%2 == 0: 
            n /=2 
            cnt +=1
        elif n%2 == 1:
            n *=3
            n +=1
            cnt +=1

```

```python
print(solution(6))
```

```python
print(solution(626331))
```
