```python
def solution(nums):
    nums_added = []
    for i in range(len(nums)-1):
        for j in range(i+1,(len(nums))):
            nums_added.append(nums[i]+nums[j])
            
    nums_added = sorted(list(set(nums_added)))
   
    return nums_added
    
```
