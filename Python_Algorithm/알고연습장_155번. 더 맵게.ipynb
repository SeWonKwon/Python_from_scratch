```python
def solution(scov, k):
    answer = 0 
    while min(scov) < k:
        scov.sort(reverse=True)
        scov.append(scov.pop() + scov.pop() * 2)
        answer+=1
        if len(scov) <=2:
            return -1
    return answer
        
```

* 효율성 문제를 위해서 sort는 버리기로 한다. 

```python
def solution(scov, k):
    answer = 0 
    while min(scov) < k:
        if len(scov) <=1:
            return -1
        
        a =min(scov)
        scov.remove(a)
        b =min(scov)
        scov.remove(b)
        scov.append(a + b*2)
        answer += 1

        
    return answer
```

heapq 사용

```python
import heapq
  
  
def solution(scoville, k):
    heap = []
    for num in scoville:
        heapq.heappush(heap, num)

    mix_cnt = 0
    while heap[0] < k:
        try:
            heapq.heappush(heap, heapq.heappop(heap) + (heapq.heappop(heap) * 2))
        except IndexError:
            return -1
        mix_cnt += 1

    return mix_cnt
```
