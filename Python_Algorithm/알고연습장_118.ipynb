함수 정의

```python
def solution(n, arr1, arr2):
    answer_list=[]
    for i, j in zip(arr1, arr2):
        a = bin(i | j)
        a = a[2:]
        a = a.zfill(n)
        a = a.replace('1', '#')
        a = a.replace('0', ' ')
        answer_list.append(a)
    return answer_list
```

```python
print(solution(n, arr1, arr2))
```
