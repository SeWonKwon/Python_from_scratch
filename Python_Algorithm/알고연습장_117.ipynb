```python
if len(s)%2==0:
    print(s[int(len(s)/2-1):int(len(s)/2+1)])
else:
    print(s[int((len(s)-1)/2)])
```

```python
def solution(s):
    if len(s)%2==0:
        answer = s[int(len(s)/2-1):int(len(s)/2+1)]
    else:
        answer = s[int((len(s)-1)/2)]
    return answer
```
