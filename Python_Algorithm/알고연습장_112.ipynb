```python
def solution(d, budget):
    d.sort()
    cnt = 0 
    for i in d:
        budget -= i 
        if budget >= 0:
            cnt+=1
        elif budget < 0:
            break
    return cnt
```

```python
d = [2,2,3,3]
budget = 10
```

```python
print(solution(d, budget))
```
