````python
from collections import OrderedDict

def add_one(d,n):
    if d[n]==0:
        d[n]=1
    elif d[n]==1:
        d[n]=2
    elif d[n]==2:
        d[n]=3
    elif d[n]==3:
        d[n]=1
        try:
            add_one(d,n+1)
        except:
            d[n+1]=0
            add_one(d,n+1)
    return d

def solution(num):
    
    d = OrderedDict({1:1})
    for _ in range(num-1):
        d = add_one(d,1)
        
#     order= sorted(list(d.keys()))
    
    answer=''
    for i in d:
        if d[i]==3:
            answer='4' + answer
        else:
            answer=str(d[i]) + answer
            
    return answer
````
