```python
def solution(s, n):
    answer=''
    for i in s:
        if ord(i)>=97: # 소문자 97~122
            if ord(i) + n > 122:
                answer += chr(96+((ord(i)+n)%122))
            else:
                answer += chr(ord(i)+n)
        elif ord(i)>=65: # 대문자 65 ~ 90
            if ord(i) + n > 90:
                answer += chr(64+((ord(i)+n)%90))
            else:
                answer += chr(ord(i)+n)
        else:
            answer += i
            
    return answer
```

```python
solution(s, n)
```

다른풀이

```python
def caesar(s, n):
    lower_list = "abcdefghijklmnopqrstuvwxyz"
    upper_list = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

    result = []

    for i in s:
        if i is " ":
            result.append(" ")
        elif i.islower() is True:
            new_ = lower_list.find(i) + n
            result.append(lower_list[new_ % 26])
        else:
            new_ = upper_list.find(i) + n
            result.append(upper_list[new_ % 26])
    return "".join(result)
```
