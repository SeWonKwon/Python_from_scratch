````python
import itertools
def solution(numbers):
    num_str = [str(i) for i in numbers]
    
    per_case=itertools.permutations(range(len(numbers)),len(numbers))
    result = []

    for i in per_case:
        num = ''
        for j in i:
            num += num_str[j]
        else:
            result.append(int(num))
        
    return str(max(result))
````
