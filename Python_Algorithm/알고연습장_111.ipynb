```python
def solution(left, right):
    answer=0 
    n = range(left, right+1)
    
    for i in n:

        divided_cnt = 0 
        num = 0 
        while True:
            num += 1
            if i == num:
                divided_cnt += 1
                break
            elif i%num == 0 :
                divided_cnt += 1
        if divided_cnt%2 == 0:
            answer += i
        elif divided_cnt%2 == 1:
            answer -= i
    
    return answer
    
```

```python
print(solution(left,right))
```

```python
left, right = 24, 27
print(solution(left,right))
```

약수의 갯수의 성질에 관한 풀이:
완전 제곱수의 약수의 갯수만 홀수개 이다. 

```python
def solution(left, right):
    answer = 0
    for i in range(left,right+1):
        if int(i**0.5)==i**0.5:
            answer -= i
        else:
            answer += i
    return answer
```

```python
left, right = 24, 27
print(solution(left,right))
```
