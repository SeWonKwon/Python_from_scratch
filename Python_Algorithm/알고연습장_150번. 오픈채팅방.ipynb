```python
def solution(record):
    Nick_name_list={}
    answer = []
    for record_ in record:
        if record_.startswith('E'):
            Nick_name_list[record_.split()[1]] = record_.split()[2]

        elif record_.startswith('C'):
            Nick_name_list[record_.split()[1]] = record_.split()[2]


    for record_ in record:
        if record_.startswith('E'):
            answer.append(f'{Nick_name_list[record_.split()[1]]}님이 들어왔습니다.')


        elif record_.startswith('L'):
            answer.append(f'{Nick_name_list[record_.split()[1]]}님이 나갔습니다.')
    return answer
```
