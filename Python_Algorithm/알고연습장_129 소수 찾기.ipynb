소수찾기

아리스토리스체를 적용해 볼까~



```python
n = 10
num_list = [0,0] + [1] * (n -1)
```

```python
num_list
```

```python
for i, j  in enumerate(num_list):
    if j == 1:
        print(i)
        num_list[2*i::i] = [0] * len(num_list[2*i::i])
        print(num_list)
```

```python
num_list
```

```python
def solution(n):
    num_list = [0, 0] + [1] * (n -1)
    for i, j in enumerate(num_list):
        if j == 1:
            num_list[2*i::i] = [0] * len(num_list[2*i::i])
    return sum(num_list)
```

```python
solution(n)
```

다른 풀이

```python
def solution(n):
    num=set(range(2,n+1))

    for i in range(2,n+1):
        if i in num:
            num-=set(range(2*i,n+1,i))
    return len(num)
```
