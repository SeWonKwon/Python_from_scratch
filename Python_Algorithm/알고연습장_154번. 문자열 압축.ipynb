함수로 한번 만들어 보자구~

```python
def solution(s):
    result = []
    if len(s)==1:
        return 1
    
    for length in range(1, len(s)//2+1):
        s_split=[]
        for i in range(len(s)//length):
            s_split.append(s[length * i : length * i + length])
        else:
            if len(s)%length != 0:
                s_split.append(s[-(len(s)%length):])
        
        new_s=[]
        cnt = 1 
        same_s =s_split[0]
        for i, j in enumerate(s_split[:-1]):
            if same_s == s_split[i+1]:
                cnt +=1
            else:
                new_s.append(str(cnt))
                new_s.append(same_s)
                same_s=s_split[i+1]
                cnt = 1
        else:
            new_s.append(str(cnt))
            new_s.append(same_s)

        new_string = ''.join([s for s in new_s if s != '1'])

        result.append(len(new_string))
        
    return min(result)
                
                
```

```python
s = "abcabcabcabcdededededede"
solution(s)
```
