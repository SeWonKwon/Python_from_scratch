# 8.1 복잡도 종류  Complexity class

**복잡도 종류**는 계산에 드는 복잡도에 따라 문제들을 분류해놓은 집합을 말한다. <br> 복잡도 **환산**(변환)<sub>reduction</sub>으로서 한 문제를 원래 문제만큼 어렵게 다른 문제로 변환하는 것을 뜻한다.

예를 들어서 <br>

\begin{equation}
a \times b = \frac {\Big({(a+b)^2-a^2-b^2}\Big)}{2}
\end{equation}

가장 일반적으로 사용되는 환산은 **다항 시간 환산**<sub>polynomial-time reduction</sub>이다. 이는 환산 절차가 다항 시간이 걸리다는 것을 의미한다.

## P

복잡도 **P**<sub>polynomial time</sub>는 결정론적 튜링 기계<sub> deterministic Turing machine</sub>로 다항 시간 안에 풀 수 있는 판정 문제<sub>decision problem</sub>를 모아 놓은 복잡도 종류다.
    

결정론적이란? 쉽게 풀어 말하자면 계산의 각 단계에서 단 한 가지 가능성만을 고려 할 수 있는 알고리즘으로 다항 시간 안에 풀 수 있는 문제다. 
    
어떤 문제를 결정론적 문제로 바꿀 수 있다면, 그 결과는 **P**에 속하게 된다.

##  NP

**NP**<sub>non-deterministic polynomial time 는 비결정론적 튜링 기계로 다항 시간 안에 풀 수 있는 판정 문제의 집합니다. 
    
P 와 구분되는 것은 비결정론적이라는 점으로 , 각 계산 단계에서 여러 가능성을 동시에 고려할 수 있는 알고리즘을 통해 다항 시간 안에 풀 수 있다는 뜻이다. 
    
어떤 문제가 적어도 NP에 속하는 어떤 문제만큼 어렵다면(다항 시간 환산으로 판정될 수 있다면) **NP-난해**<sub>NP-hard</sub>이라 한다.
    
다른 정의로는, 모든 NP 문제가 다항 시간에 어떤 문제로 환산될 수 있다면, 그 문제를 NP-난해라고 한다. 
        <br>예로, 그래프를 통해 최단 경로를 찾는 외판원 문제는 NP-난해에 속한다.

NP-난해 문제 자체는 NP 문제가 아닐수도 있다.( NP보다 더 어려운 문제일 수도 있다. ) NP 문제이자 NP-난해인 문제를 **NP-완전**<sub>NP-complete</sub> 이라고 부른다. 
    
<br>NP-완전은 NP 집합에 속하는 판정 문제 중에서 가장 어려운 문제의 부분집합으로, 모든 NP 문제는 다항 시간 내에 NP-완전 문제로 환산할 수 있다.

### P와 NP
<br>
NP에 속하는 문제는 그 문제의 대답이 '예'라는 근거가 주어졌을 때 그것이 옳은 근거임을 다항식 시간에 확인해줄 수 있는 문제다. 반면, P에 속하는 문제는 그 문제의 질문이 주어졌을 때, Yes 또는 No의 대답을 다항식 시간에 해줄 수 있는 문제다. NP 문제는 대답이 Yes라는 근거가 제시되었을 때 그것이 맞는다는 것을 다항식 시간에 확인만 하면 될 뿐 대답이 No일 경우는 어떻게 하든 상관없다. P 문제는 질문 자체에 대해서 다항식 시간에 Yes 또는 No 대답을 해야 한다. 질문에 대답을 한다는 것은 일반적으로 '문제를 푼다'는 것과 일치한다.



예를 들어


> [문제 1] 양의 정수 X는 3의 배수인가?
>
> [변환] X가 n자리의 수라면 $$ X = x_n \times 10^{n-1} + x_{n-1} \times 10^{n-2} + \cdots +x_1$$로 나타낼 수 있다.<br> 각 자릿수를 다음과 같이 단순히 더한 수 Y를 만든다. $$ Y = x_n + x_{n-1} + \cdots + x_1 $$
> [문제 2] Y는 3의 배수인가?

[문제 1]의 답은 [문제 2]의 대답과 일치한다. 

문제 1을 풀고 싶으면 문제 2로 변환을 해서 문제 2의 답(Yes 또는 No)을 문제 1의 답으로 삼으면 된다. 문제 2를 쉽게 풀 수 있으면 문제 1도 쉽게 풀 수 있다. 

해결해야 할 문제 B가 어느 정도 어려운지 모른다고 하자. 유명한 난제군인 NP-완전에 속하는 어떤 문제 A를 문제 1의 위치에 놓고 적당히 변환해 문제 B로 바꾸고 문제 B가 문제2와 같이 문제 1과 대답이 일치한다면 문제 B는 NP-난해 문제, 즉 이 유명한 나제군에 속한다.

## P=NP ?

**P-NP 문제**<sub> P versus NP problem </sub> 는 복잡도 종류 P와 NP가 같은지 여부에 대한 컴퓨터 과학 분야의 미해결 문제다. <br>

앞에서 언급했듯이 P는 결정론적 튜링 기계로, NP는 비결정론적 튜링 기계로 다항 시간에 답을 구할 수 있는 문제의 집합이다. 
    
결정론적 튜링 기계에 사용한 프로그램은 비결정론적 튜링 기계에도 적용할 수 있으므로, P는 NP의 부분집합이 된다. 하지만 P와 NP가 같은 집합인지, 아니면 P가 NP의 진부분집합인지 여부는 아직 밝혀지지 않았다. 

NP- 문제 중 하나라도 P에 속한다는 것을 증명한다면 모든 NP 문제가 P에 속하기 때문에, P-NP 문제가 P=NP 가 되고 반대로 NP-완전 문제 중 하나가 P에 속하지 않는다는 것이 증명된다면 $ P=NP$에 대한 반례가 되어 $ P \neq NP$가 된다.

#  재귀 알고리즘

**재귀 알고리즘**<sub>recursion algorithm<sub/>의 3가지 법칙은 다음과 같다. 

1. 재귀 알고리즘은 **베이스 케이스**<sub>base case</sub> 가 있어야한다. 
    - 베이스 케이스란? 재귀 호출을 유발하지 않는 종료 시나리오<br><br>
2. 재귀 알고리즘은 상태를 변경하고 베이스 케이스로 이동한다.    
<br>    
3. 재귀 알고리즘은 재귀적으로 자신을 호출한다. 

모든 재귀 호출에 대해 재귀 함수는 인수, 반환 주소, 지역 변수를 메모리의 스택에 할당한다. <br> 이러한 데이터를 스택에 넣고<sub>push</sub> 꺼내느<sub>pop</sub>데에는 시간이 소비된다. 

재귀 알고리즘은 최소 $O(n)$의 공간을 사용한다. n 은 재귀 호출의 깊<sub>depth</sub>이다.

재귀는 계산이 중복되거나 하위 문제가 겹치는 경우 비용이 많이 든다. 어떤 경우에는 스택 오버플로가 발생할수도 있다.<br>
이러한 이유로 하위 문제가 겹치는 경우에는 반복문을 사용하는 것이 더 좋은 방법이 될 수도 있다. 

예를 들면 피보나치 수열의 경우 단순히 재귀 함수를 사용한다면 지수 시간 $O(2n)$이 걸리지만, 반복문을 사용한다면 $O(n)$이 걸린다.

##  재귀식

**재귀식**<sub>recursive relation</sub>

$$T(n) = a \cdot T(g(n)) + f(n)$$

a는 재귀 호출 수를 나타낸다. $g(n)$은 재귀적으로 풀어야 할 각 하위 문제의 크기를 나타낸다. $f(n)$은 함수에서 수행되는 모든 추가 작업이다.

## 분할정복 알고리즘

**분할정복**<sub>divide and conquer</sub> 알고리즘의 재귀식

$$ T(n) = a \cdot T( \frac {n} {b}) + f(n) $$

a 는 재귀 호출 수를 나타내고, 각 호출은 전체 데이터에서 $\frac {1} {b} $ 만큼을 사용하면 $f(n)$의 작업을 수행한다. <br>

마지막 사례(말단 노드)에 도달했을 때 $T(1) = 1$이 되기 위해 **높이**<sub>height</sub> $h=log_b{n}$으로 정의한다.

# 함수의 실행 시간 

알고리즘의 실행 시간(런타임)<sub>runtime</sub><br>

알고리즘에 재귀 호출이 없다면, 자료구조와 코드 블록 흐름만 파악하면 된다. 

* 하나씩 실행되는 코드 블록의 시간복잡도를 계산하면 된다. 
* 중첩 반복문의 경우 시간복잡도는 배가 된다. 

* 알고리즘에 재귀 호출이 있다면 재귀식을 사용하여 실행 시간을 찾을 수 있다. 
    - 함수에 대한 재귀식을 작성할 때는<br> *베이스 케이스*($O(1)$이어야 하므로 $T(1)=1$)와 *일반 케이스* 두 방정식을작성해야 한다. 
    
    

## 피보나치 수열에서 n번째 요소를 찾는 알고리즘 예제
피보나치 수열에서 단순 재귀 함수 알고리즘의 시간복잡도는 지수 시간이 걸린다. 

```python
def find_fibonacci_seq_rec(n):
    if n < 2:
        return n
    return find_fibonacci_seq_rec(n-1) + find_fibonacci_seq_rec(n -2)
```

```python
print(find_fibonacci_seq_rec(10))
```

**피보나치 수열**

$$ a_n = a_{n-1} + a_{n-2} $$

**재귀식**<sub>recursive relation</sub>

$$T(n) = a \cdot T(g(n)) + f(n)$$

a는 재귀 호출 수를 나타낸다. $g(n)$은 재귀적으로 풀어야 할 각 하위 문제의 크기를 나타낸다. $f(n)$은 함수에서 수행되는 모든 추가 작업이다. 

* $g(n)$ n-2와 n-1
* $a$은 2
* $f(n)$ 은 1

다음 단계의 재귀식들은 다음과 같이 될것이다. 

$$ T(n) = 2^{2} \cdot T(n-2) + 2$$

즉,
$$ T(n) = 2^{k} \cdot T(n-k) + k \dots $$

베이스 케이스는 $ T(1) = 1 $ 이며, 시간 복잡도는 O(1)이 되어야 하므로, $n - k = 1$ ,<br> 즉 $k = n - 1$이다

이를 대입하여 재귀식을 정리하면 다음과 같다. 

$$ T(n) = 2^{n-1} + n-1 \sim 2^n $$

이렇듯 피보나치 수열 알고리즘의 시간복잡도가 지수 시간인 $O(2^n)인 것을 실제로 증명해봤다. <br>

동일한 과정으로 다른 문제의 재귀식도 세울수 있다.(정확한 계산이라기보다는 재귀식을 통해 실행 시간을 추론하는 것이다. )
