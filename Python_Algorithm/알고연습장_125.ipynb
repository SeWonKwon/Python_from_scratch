* 영어 알파벳의 재정렬
* 대문자가 뒤로
* 즉, decsending

```python
s = 'Zbcdefg'
```

```python
new_s_low=[]
new_s_up=[]
for i in s:
    if i == i.lower():
        new_s_low.append(i)
    else:
        new_s_up.append(i)

print(new_s_low, new_s_up)
```

```python
answer = ''
for i in reversed(new_s_low):
    answer += i
for i in reversed(new_s_up):
    answer += i

print(answer)
```

함수 정의

```python
def solution(s):
    new_s_low=[]
    new_s_up=[]
    for i in s:
        if i == i.lower():
            new_s_low.append(i)
        else:
            new_s_up.append(i)
    answer = ''
    for i in (sorted(new_s_low))[::-1]:
        answer += i
    for i in (sorted(new_s_up))[::-1]:
        answer += i
    
    return answer
```
