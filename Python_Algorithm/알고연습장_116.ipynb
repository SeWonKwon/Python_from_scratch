```python
def solution(a, b):
    Day_list = ['SUN' ,'MON', 'TUE', 'WED', 'THU', 'FRI', 'SAT']
    Month_daylist = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    a2 = sum(Month_daylist[:a-1])
    return Day_list[ (b + a2 + 4)%7 ]
```
