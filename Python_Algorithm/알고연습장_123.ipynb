```python
def solution(strings, n):
    new_s = {}
    for i in strings:
        try:
            temp = new_s[i[n]]
            temp.append(i)
            new_s[i[n]]= sorted(temp)
        except:
            new_s[i[n]]= [i] 
    
    answer = []
    for i in sorted(new_s.keys()):
         for j in new_s[i]:
                answer.append(j)
    return answer
```
