우선 3진법으로 바꾼후에 , 그 값에 따라서 1, 2, 4 로 리플레이스 해주기로 한다. 

```python
n = 15
```

```python
n, r=divmod(n, 4)
print(n, r)
```

```python
num_4 = ''
n = 15
while True:
    n, r = divmod(n, 4)
    
    if n == 0 and r == 0:
        break
        
    num_4 = str(r) + num_4

print(num_4)
```

```python
num_4 = num_4.replace('3','4')
```

```python
num_4
```

```python
def solution(n):
    num_4 = ''
    while True:
        n, r = divmod(n, 3)

        if n == 0 and r == 0:
            break

        num_4 = str(r) + num_4
        num_4 = num_4.replace('10','4')
        num_4 = num_4.replace('20','14')
        num_4 = num_4.replace('40','24')
    return num_4
```

```python
for i in range(10):
    print(i, solution(i))
```

```python
def solution(n):
    i = 0
    num_124 = ''
    while True:
        a = n // 3**i
        if a == 0 :
            break
        num_124 = num_124 + str(a)
        i +=1
    num_124 = num_124.replace('3', '4')
    return num_124
```

```python
for i in range(10):
    print(i, solution(i))
```
