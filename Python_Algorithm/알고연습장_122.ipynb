* a와 b 사이의 수의 합
* 대소 관계 없음

```python
a, b = 3, 3
```

```python
a, b = 5, 3
```

```python
a, b  = min(a,b), max(a,b)
```

```python
print(a, b)
```

```python
sum(range(a, b+1))
```

```python
def solution(a,b):
    a, b  = min(a,b), max(a,b)
    return sum(range(a, b+1))
```
