```python
s = 'try hello world'
```

```python
def solution(s):
    answer=''
    j = 0
    for i in s:
        if i == ' ':
            j = 0
            answer+=i
            
        elif j%2==0:
            answer+=i.upper()
            j+=1
            
        else:
            answer+=i.lower()
            j+=1
    return answer
```

```python
s = 'aa  a s d fasdf'
```

```python
solution(s)
```
