```python
def solution(nums):
    nums_added = []
    cnt = 0 
    for i in nums:
        for j in nums:
            for k in nums:
                if i == j or j==k or i ==k :
                    pass
                else:
                    nums_added.append(i+j+k)
    
    for i in nums_added:
        n = 1
        while True:
            n+=1
            if i==n :
                cnt+1
            elif i % n == 0 : 
                break
    return cnt
```
