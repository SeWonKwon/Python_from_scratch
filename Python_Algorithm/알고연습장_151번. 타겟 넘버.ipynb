```python
def solution(numbers, target):
    import numpy as np
    n = len(numbers)
    add_to_l = [[1],[-1]]
    ndim=0 
    
    while ndim != n-1 :
        ndim+=1
        for i in add_to_l[:]:
            add_to_l.remove(i)
            add_to_l.append(i+[1])
            add_to_l.append(i+[-1])
            
    cnt = 0 
    for i in add_to_l:
        if sum(np.array(numbers)*np.array(i)) == target:
            cnt+=1
    return cnt
```

```python
def solution(numbers, target):
    result = 0
    def dfs(num, level):
        nonlocal result

        if level == len(numbers):
            if num == target:
                result += 1
            return

        signs = [-num, num]
        if level == 1:
            for i in range(2):
                dfs(signs[i] + numbers[level], level + 1)
                dfs(signs[i] - numbers[level], level + 1)
        else:
            dfs(num + numbers[level], level + 1)
            dfs(num - numbers[level], level + 1)

    dfs(numbers[0], 1)
    return result
```

```python
from itertools import product
def solution(numbers, target):
    l = [(x, -x) for x in numbers]
    s = list(map(sum, product(*l)))
    return s.count(target)
```

```python
def solution(numbers, target):
    if not numbers and target == 0 :
        return 1
    elif not numbers:
        return 0
    else:
        return solution(numbers[1:], target-numbers[0]) + solution(numbers[1:], target+numbers[0])
```
