```python
def gcd(a, b):
    return b if a % b == 0 else gcd(b, a % b)

def solution(w, h):
    gcd_ = gcd(w, h)
    w1, h1 = w/gcd_, h/gcd_
    return w * h - gcd_ * (w1 + h1 -1)
    
```
