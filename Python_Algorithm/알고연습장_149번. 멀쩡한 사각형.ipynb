분석해보면 , 최대 공약수와 연관이 있어 보인다. 
위의 그림을 보면 
- 가로가 즉 w 는 2 * 4 
- 세로가 즉 h 는 3 * 4
2 * 3 에 대각선이 4개가 그려지는 형태임을 알수 있다. 

2 * 3 의 상자는 6개 중에 4개를 못 쓰게 되는것이다. 

즉, 변과 직선이 만나는 점,  점을 만나게 되는 갯수

우선 최대공약수 함수 정의

```python
def gcd(a, b):
    return b if a % b == 0 else gcd(b, a % b)
```

```python
gcd(8, 12)
```

즉 점을 지나지 않기 때문에 , 맨하탄 거리와 같다고 볼수 있다. 

```python
def gcd(a, b):
    return b if a % b == 0 else gcd(b, a % b)

def solution(w, h):
    gcd_ = gcd(w, h)
    w1, h1 = w/gcd_, h/gcd_
    return w * h - gcd_ * (w1 + h1 -1)
    
```
