우선은 정확도 테스트에서 빠진 부분을 찾아 냈다. 

아마도 마지막에 어펜드가 떨그렁 되는게 아닌가 싶다. 

그래서 전부 어펜드를 시켰더니 효용성에서 실패가 뜨게 된다. 

```python
def solution(s):
    answer=[]
    for i in s:
        if len(answer)==0:
            answer.append(i)
        elif answer[-1] == i:
            answer= answer[:-1]
        else:
            answer.append(i)
#         print(answer)

    if answer==[]: return 1
    else: return 0
```
