## 매직 명령어(magic commands)

* Jupyter Notebook 환경에서 파이썬 파일 저장 및 실행

````python
%%writefile test.py
print('magic command')
````

````python
%run test.py
````

* 코드 실행 시간 측정
  - % 가 한개 일때는 단일문
  - %% 여러줄의 코드문

````python
%%timeit?
````

````python
%timeit li = [n ** n for n in range(1000)]
````

````python
%%timeit
li = []
for n in range(1000):
  li.append(n ** n)
````

## 입력과 출력 이력

* In과 Out 객체를 통해 이전 명령어와 명령어의 출력 결과 보기

  - In : 내린 명령어 조회
  - Out : 나온 결과물
  - +[index] : 앞에 나온 결과물

````python
In[6]
````

````python
Out[6]
````

````python
Out[18]
````

````python
print("In[18]:" + str(In[18] )+ "\nOut[18]: " + str(Out[18]))
````

* `_`를 이용해 이전 출력값 출력하기

````python
# 이전
print(_)
````

````python
# 이전 이전
print(__)
````

````python
# 이전 이전 이전
print(___)
````

````python
_18
````
