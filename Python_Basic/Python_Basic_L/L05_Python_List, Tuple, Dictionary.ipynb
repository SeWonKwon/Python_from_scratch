## 리스트(List)

* 변경 가능한 시퀀스 자료형
* 하나의 변수에 여러 값 할당 가능
* 같은 자료형이 아니라 다른 자료형을 가지는 값들을 포함 가능
* 일반적으로 비슷한 항목들의 모음을 순서대로 저장하는데 사용




image.png

```python
print([])
print([1, 2, 3])
print(['one', 'two', 'three'])
print([1, 'One', 2, 'Two', 3, 'Three'])
print([1, 2, 3, ['One', 'Two', 'Three']])
```

### 리스트 인덱싱(List Indexing)

* 리스트의 각 위치에 해당하는 값에 접근하기 위해서 주소 개념의 숫자를 사용

image.png

```python
l = ['One', 'Two', 'Three']
print(l)
print(l[0])
print(l[1])
print(l[2])
print(l[-1])
print(l[-2])
print(l[-3])
```

### 중첩 리스트 인덱싱(Nested List Indexing)

* 리스트 안에 리스트가 있을 경우 인덱스 접근 방법

image.png

```python
Nested_list = [1, 2, 3, ['One', 'Two', 'Three']]
print(Nested_list)
print(Nested_list[3][0])
print(Nested_list[3][1])
print(Nested_list[3][2])
```

### 리스트 슬라이싱(List Slicing)

* 리스트의 인덱스를 통해서 부분만 가져올 때 사용

```python
list = ['One', 'Two', 'Three']
print(list)
print(list[0:])
print(list[1:2])
print(list[1:])
```

### 중첩 리스트 슬라이싱(Nested List Slicing)

* 중첩된 리스트에서 일부분만 인덱스를 통해서 가져올 때 사용

```python
Nested_list = [1, 2, 3, ['One', 'Two', 'Three']]
print(Nested_list[2:4])
print(Nested_list[3][0])
print(Nested_list[3][1:])
print(Nested_list[3][:2])
```

### 리스트 연산자/함수(List Operators/Function)

* 더하기 + 연산자
* 곱하기 * 연산자
* 리스트 길이를 구하는 `len()` 함수

```python
list_1 = ['One', 'Two', 'Three']
list_2 = ['Four', 'Five', 'Six']
print(list_1)
print(list_2)
print(list_1+ list_2)
print(list_1 * 3)
print(len(list_1))
print(len(list_1 * 3))
```

### 리스트 수정(List Modify)

* 리스트 인덱스 주소를 이용한 값 수정

```python
list = ['One', 'Two', 'Three']
print(list)
list[2] = 'Sam'
list
```

### 리스트 언패킹

**시퀀스 언패킹 연산자**<sub>sequence unpacking operator</sub>

모든 반복 가능한<sub>iterable</sub> 객체는 `*`를 사용하여 언패킹 할수 있다. 

```python
x, *y, z = [1, 2, 3, 4, 5, 6]
y
```

### 리스트 메소드(List Methods)


#### append()

* 리스트 요소 추가

```python
l = ['One', 'Two', 'Three']
l.append('Four')
l
```

#### sort()

* 리스트 정렬

```python
l = [10, 40, 20, 30]
print(l)
l.sort()
print(l)
l.sort(reverse=True)
print(l)
l = ['orange', 'apple', 'banana', 'strawberry']
print(l)
l.sort()
print(l)
l.sort(reverse=True)
print(l)
```

l.sort() vs sorted(l)
* `l.sort()` 는 l를 정렬하는 함수
* `sorted(l)` 는 정렬된 l를 return 하는 함수

```python
print(sorted(l))
```

#### reverse()

* 리스트 요소 반전

```python
l = ['orange', 'apple', 'banana', 'strawberry']
l.reverse()
print(l)
```

`reversed(l)`

```python
l = ['orange', 'apple', 'banana', 'strawberry']
print(l)
print(reversed(l))
for i in reversed(l):
    print(i)
```

#### index()

* 리스트의 요소 값에 대한 인덱스를 반환

```python
l = [10, 40, 20, 30]
print(l)
print(l.index(10))
print(l.index(20))
```

#### insert()

* 리스트 요소 삽입

    - `append()`는 마지막에, `insert()` 내가 원하는 위치에

```python
l = [10, 40, 20, 30]
print(l)
l.insert(4, 50)
print(l)
l.insert(0, 60)
print(l)
```

#### remove()

* 리스트 요소 제거

```python
l = [10, 40, 20, 30]
l.remove(40)
print(l)
```

#### del

* 리스트 요소 제거 연산

```python
l = [10, 40, 20, 30]
print(l)
del l[1]
print(l)
del l[2]
print(l)
```

`remove()` vs `del()`
* `remove()` :  값으로 제거.
* `del()` : 인덱스로 제거.

#### pop()

* 리스트 요소를 방출

```python
l = [10, 30, 40, 20]
l.pop()
```

```python
print(l)
l.pop(2)
```

#### count()

* 리스트 요소의 갯수 계산

```python
l = [10, 20, 20, 30, 30, 30]
print(l)
print(l.count(30))
print(l.count(20))
```

#### extend()

* 리스트 확장

* `list.append(x)`는 리스트 끝에 x 1개를 그대로 넣습니다.

* `list.extend(iterable)`는 리스트 끝에 가장 바깥쪽 iterable의 모든 항목을 넣습니다.

```python
l = [10, 40, 20, 30]
print(l)
l.extend([50, 60])
l
```

## 튜플(Tuple)

* 리스트와 유사하지만 변경 불가능한 시퀀스 자료형
* 하나의 변수에 여러 값 할당 가능
* ‘(‘와 ‘)’를 사용하여 표현
*  괄호를 쓰지 않으면 튜플이다.


```python
print(())
print((1, 2, 3))
print(('One', 'Two', 'Three'))
print((1, 'One', 2, 'Two', 3, 'Three'))
print((1, 2, 3, ('One', 'Two', 'Three')))
```

* 한개 짜리 튜플을 만드는 법

`a = (30,)`


`a = 30,` 

```python
a = (30,)
b = 30
print(type(a), a)
print(type(b), b)
```

```python
a  = 1, 2, 3,

print(type(a), a)
```

### 튜플 슬라이싱(Tuple Slicing)

## 세트(Set)

* 데이터 중복을 허용하지 않는 구조
* 순서가 없는 데이터 집합을 위한 구조
* 인덱싱으로 값을 접근할 수 없음

```python
print({})
print({'One', 'Two','Three'})
print({10, 20, 30, 40})
```

### 세트 연산자(Set Operators)

* 교집합: &
* 합집합: |
* 차집합: -
* 여집합: ^ ( 교집합의 여집합)

```python
s_1 = {10, 20, 20, 30}
s_2 = {30, 30, 40, 50}

print(s_1, s_2)
print(s_1 & s_2)
print(s_1 | s_2)
print(s_1 - s_2)
print(s_1 ^ s_2)
```

### 세트 메소드(Set Methods)

* 교집합: intersection()
* 합집합: union()
* 차집합: difference()
* 여집합: symmetric_difference()

```python
s_1 = {10, 20, 20, 30}
s_2 = {30, 30, 40, 50}
print(s_1, s_2)

print(s_1.intersection(s_2))
print(s_1.union(s_2))
print(s_1.difference(s_2))
print(s_1.symmetric_difference(s_2))
```

* 요소 추가(한개): add()
* 여러 요소 추가: update()
* 요소 제거: remove()
* 요소 제거: discard()
* 모든 요소 제거: clear()

```python
s = {10, 20, 30, 40}
print(s)

s.add(50)
print(s)

s.update([60, 70])
print(s)

s.remove(70)
s.remove(60)
print(s)

s.discard(30)
print(s)

s.clear()
print(s)
```

## 딕셔너리(Dictionary)

* 키(key)와 값(value)의 쌍으로 구성된 데이터
* 순서가 없는 데이터
* 키를 통해 값을 얻음
* 동일한 키가 있을경우 덮어씀( 키는 유일함 )

```python
dic = {1: 'One', 2: 'Two', 3:'Three'}
print(dic)

dic = {1: 'One', 2: 'Two', 3:'Three',1: 'One', 2: 'Two', 3:'Three'}
dic
```

### 딕셔너리 요소 추가/삭제

* 딕셔너리의 해당 키 값에 값을 추가하여 요소 추가
* del을 이용하여 요소 제거

```python
dic = {1: 'One', 2: 'Two', 3:'Three'}
print(dic)
print(dic[2]) # index 로 접근한 것이 아니라 key값으로 접근함
dic[4] = 'Four' # key 와 value 추가
print(dic)
dic[5] = 'Five'
print(dic)
del dic[4] # 제거
print(dic)
```

### 딕셔너리 메소드(Dictionary Methods)


|메소드|설명|
|---|---|
|keys()|딕셔너리의 키 가져오기|
|values()| 딕셔너리의 값 가져오기|
|items()|딕셔너리의 키와 값을 모두 가져오기|
|get()|딕셔너리에서 키에 해당하는 값 가져오기|
|pop()|딕셔너리에서 키에 해당하는 값 추출하기메소드|
|clear()|딕셔너리의 모든 요소를 제거하기|






```python
dic = {1: 'One', 2: 'Two', 3:'Three'}
print(dic)
print(dic.keys())
print(dic.values())
print(dic.items())
print(dic.get(2)) # return
print(dic.pop(3)) # return & del
print(dic)
dic.clear() # 빈 딕셔너리 return
print(dic)
```
