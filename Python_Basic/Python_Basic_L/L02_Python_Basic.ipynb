## 파이썬 소개(Python Introduction)
image.png

* 플랫폼 독립적이며 인터프리터식, 객체지향적, 동적 타이핑(dynamictyped) 대화형 언어
* 파이썬이라는 이름은 귀도 반 로섬 (1989) 가 좋아하는 코디미 <Monty Python's Flying Circus>에서 따온 것
* 파이썬의 사전적인 의미는 비단뱀으로 로고 디자인도 파랑색과 노란색 비단뱀 두 마리가 서로얽혀 있는 형태

### 파이썬 특징(Python Features)
1. 쉬운 문법과 간결함
    * 사람의 사고 체계와 유사하고 문법 자체가 아주 쉽고 간결함
    * 소스 코드가 이해하기 쉬워 작업과 유지 보수가 쉬움
2. 동적 타이핑(dynamic typing) 범용 프로그래밍 언어
    * 프로그램의 실행 시점에서 프로그램 변수의 타입을 결정하는 언어
3. 다양한 플랫폼 지원과 호환성
    * 다양한 플랫폼에서 동작하도록 지원
    * C언어와의 호환성이 매우 뛰어남
4. 풍부한 라이브러리(모듈)
    * 다양한 파이썬 내부 라이브러리 뿐만 아니라 다양한 서드 파티third party 라이브러리 사용 가능
5. 풀 언어(glue language)
    * 다른 언어로 쓰인 모듈들을 연결하는 언어로 자주 이용
    * C나 C++ 언어로 만든 프로그램을 파이썬에서 사용 가능
6. 다양한 언어의 문자 처리
    * 유니코드 문자열을 지원해서 다양한 언어의 문자 처리 가능

### 들여쓰기(Indentation)

* 코드 라인의 시작 부분에 들여쓰기
* 파이썬은 들여쓰기로 코드의 블록을 표시

```python
if 2 > 1:
    print(2)
```

```python
# if 2 > 1:
# print(2)
```

### 주석(Comments)

* 코드에 대해 설명하는데 사용하고 가독성 증가
* 일부 코드를 주석처리하여 실행 방지 가능

```python
# 주석 부분
print('주석문') # 주석문 설명을 위한 주석
```

```python
%%timeit 
print("Hello World!")
```
